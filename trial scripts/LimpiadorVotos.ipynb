```python
import pandas as pd
import numpy as np
```

Alcaldes 2018

```python
temp_content = pd.read_csv('NL_AYUN_2018.csv')
```

```python
group = temp_content.groupby('MUNICIPIO')
```

```python
NL_AYUN_2018 = group.sum()
```

```python
NL_AYUN_2018.head()
```

```python
NL_AYUN_2018.drop(['ID_ESTADO','ID_MUNICIPIO','SECCION','ID_CASILLA','EXT_CONTIGUA','UBICACION_CASILLA','TIPO_ACTA','SHA','FECHA_HORA_ACOPIO','FECHA_HORA_VERIFICACION','ORIGEN','DIGITALIZACION'

], axis = 1, inplace= True)
```

```python
NL_AYUN_2018.head()
```

```python
NL_AYUN_2018.to_csv('final_NL_AYUN_2018.csv')
```
