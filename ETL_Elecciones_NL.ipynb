## ID Municipios

```python
ID_Municipios_2015 = pd.read_csv("data/ID-Municipios-2015.csv", encoding="latin-1")
print(ID_Municipios_2015.shape)
ID_Municipios_2015.head()
```

```python
ID_Municipios_2018 = pd.read_csv("data/ID-Municipios-2018.csv", encoding="latin-1")
print(ID_Municipios_2018.shape)
ID_Municipios_2018.head()
```

## Ayuntamientos 2015

```python
ayuntamientos2015 = pd.read_csv("data/CEENL-Ayuntamientos-2015.csv", encoding="latin-1")
print(ayuntamientos2015.shape)
ayuntamientos2015.head()
```

```python
ayuntamientos2015.iloc[:,8:].columns
```

```python
votos_a2015 = ayuntamientos2015[['Municipio','Total','Partido Acción Nacional',
 'Partido Revolucionario Institucional',
 'Partido de la Revolución Democrática',
 'Partido del Trabajo',
 'Partido Verde Ecologista de México',
 'Movimiento Ciudadano',
 'Nueva Alianza Partido Político Nacional',
 'Partido Demócrata',
 'Partido Cruzada Ciudadana',
 'MORENA',
 'Partido Humanista  Partido Político Nacional',
 'Partido Político Nacional  Encuentro Social',
 'Candidato Independiente','Votos Anulados']].groupby('Municipio').sum()

print(votos_a2015.shape)
votos_a2015.head()
```

```python
ganadores_a2015 = votos_a2015.iloc[:,1:].idxmax(axis=1)
ganadores_a2015.to_list()[0:5]
```

```python
ganadores_a2015.value_counts()
```

```python
votos_a2015.insert(1, 'Ganador', ganadores_a2015)
print(votos_a2015.shape)
votos_a2015.head()
```

```python
resultados_a2015 = ID_Municipios_2015.merge(votos_a2015, on='Municipio')
print(resultados_a2015.shape)
resultados_a2015
```
